```python
from google.colab import drive
drive.mount('/content/drive')
import gradio as gr
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image

# Load the CNN model from .keras format
model = load_model('/content/drive/MyDrive/MINI_Project1B/lung_cancer_model.keras')

# Class labels
class_labels = {
    0: 'Adenocarcinoma',
    1: 'Large Cell Carcinoma',
    2: 'Normal Lung',
    3: 'Squamous Cell Carcinoma'
}

# Preprocess the uploaded image
def preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))  # Resize to model input size
    img_array = image.img_to_array(img) / 255.0  # Normalize
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    return img_array

# Make prediction and format results
def predict_lung_cancer(img):
    img_array = preprocess_image(img)

    # Get model prediction
    prediction = model.predict(img_array)
    predicted_class = np.argmax(prediction)
    confidence = round(100 * np.max(prediction), 2)

    # Create a results DataFrame
    result_df = pd.DataFrame({
        "Class": [class_labels[predicted_class]],
        "Confidence (%)": [confidence]
    })

    return result_df

# Define Gradio Interface
with gr.Blocks(theme=gr.themes.Soft()) as interface:
    gr.Markdown("<h2 style='text-align: center;'>Lung Cancer Classification</h2>")
    gr.Markdown("<p style='text-align: center;'>Upload a CT scan to predict the type of lung cancer.</p>")

    with gr.Row():
        img_input = gr.Image(type="filepath", label="Upload CT Scan")
        result_output = gr.Dataframe(label="Prediction Result", headers=["Class", "Confidence (%)"])

    # Loading button and prediction
    submit_button = gr.Button("Analyze Image")
    submit_button.click(fn=predict_lung_cancer, inputs=img_input, outputs=result_output)

# Launch Gradio UI
interface.launch(debug=True)
```
